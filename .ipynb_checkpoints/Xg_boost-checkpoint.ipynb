```python
ccdata_new=pd.read_csv('df1 (1).csv')
```

```python
ccdata_new
ccdata_new.drop(columns=['Unnamed: 0'])
```

```python
df_=ccdata_new[['Date','Prices','Comp','Negative','Neutral','Positive']].copy()
```

```python
df_
```

```python
data = df_.filter(['Prices'])
dataset = data.values
training_data_len = math.ceil( len(dataset) * .8 )

training_data_len
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(dataset)

scaled_data
```

```python
train_data = scaled_data[0:training_data_len , :]
```

```python
x_train = []
y_train = []
for i in range(60, len(train_data)):
  x_train.append(train_data[i-60:i, 0])
  y_train.append(train_data[i, 0])
  if i<= 61:
    print(x_train)
    print(y_train)
    print()
```

```python
data_ = df_.filter(['Comp'])
dataset_ = data_.values
```

```python
import numpy as np
```

```python
for i in range(60,627):
  x=np.array(dataset_[i])
  x_train[i-60]=np.append(x_train[i-60],x)
```

```python
import xgboost as xg
from sklearn.metrics import mean_squared_error as MSE
```

```python
xgb_r = xg.XGBRegressor()
```

```python
xgb_r.fit(x_train, y_train)
```

```python
test_data = scaled_data[training_data_len - 60: , :]
#Create the data sets x_test and y_test
x_test = []
y_test = dataset[training_data_len:, :]
```

```python
test_data = scaled_data[training_data_len - 60: , :]
x_test = []
y_test = dataset[training_data_len:, :]
```

```python
for i in range(60, len(test_data)):
  x_test.append(test_data[i-60:i, 0])
```

```python
for i in range(627,783):
  x=np.array(dataset_[i])
  x_test[i-627]=np.append(x_test[i-627],x)
```

```python
x_test = np.array(x_test)
```

```python
pred = xgb_r.predict(x_test)
```

```python
pred
```

```python
pred = np.reshape(pred,(-1,1))
```

```python
pred = scaler.inverse_transform(pred)
```

```python
pred
```

```python
mse=np.mean(((pred- y_test)**2))
mse
```

```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

valid = data[training_data_len:]
valid['Prediction'] = pred
#Visualize the data
plt.figure(figsize=(16,8))
plt.title('Model')
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price USD ($)', fontsize=18)

plt.plot(valid[['Prices', 'Prediction']])
plt.legend(['Val', 'Prediction'], loc='lower right')
plt.show()
```
